```python
def open_file(file):
    doc_file = open(file, 'r', encoding='ANSI')
    content = doc_file.read()
    doc_file.close()
    return content
```

```python
a = open_file("서비스별 태그 목록.txt")
```

```python
a
```

```python
corpus = a.split("\n\n")
```

```python
docs = []
for i in corpus:
    doc = dict()
    para = i.split(" / ")
    para
    doc["title"]=para[0]
    doc["content"]=para[1]
    doc["tag"]=para[-1]
    docs.append(doc)
```

```python
docs
```

```python
from konlpy.tag import Kkma
```

```python
def tokenizing(content): #입력 문장 리스트를 토크나이징한다. Kkma는 pos로 분리하고, OS, NNG, NP만 빼내는 것이 좋다.
    kkma = Kkma()
    featurelist = []
    for term in kkma.pos(content):
        if term[1] == 'OS':
            featurelist.append(term[0])
        elif term[1] == 'NNG':
            featurelist.append(term[0])
        elif term[1] == 'NP':
            featurelist.append(term[0])
        elif term[1] == 'OL':
            featurelist.append(term[0])
        elif term[1] == 'NL':
            featurelist.append(term[0])
    return featurelist
```

```python
for doc in docs:
    doc["tokens"]=tokenizing(doc["content"])+tokenizing(doc["tag"])
```

```python
from scipy.spatial import distance
def searching(term, d2v): #단어와 각 문서간의 유클라디안 거리를 계산하는 함수
    d2v_dict = dict()
    for i in range(d2v.docvecs.count):
        d2v_dict[d2v.docvecs.index_to_doctag(i)] = distance.euclidean(d2v[term],d2v.docvecs[i])
        #print(d_key+" : "+str(distance.euclidean(d2v[term],d2v.docvecs[i])))
    return d2v_dict
```

```python
d2v_rank = searching('IT',new_d2v_model) #각 문서와 단어간에 유클라디안을 계산하여 저장
for i in sorted(d2v_rank, key=lambda k : d2v_rank[k]): #기술에 대한 각 문서별 유클라디안 거리를 오름차순으로 정렬한 모습.
    print( i + " : " + str(d2v_rank[i]))
```

```python
from sklearn.manifold import TSNE #단어 분포를 보여주기 위한 차원 축소 패키지
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
def plot_with_labels(low_dim_embs, labels, filename='tsne.png'):
    assert low_dim_embs.shape[0] >= len(labels), "More labels than embeddings"
    plt.figure(figsize=(18, 18))  # in inches
    for i, label in enumerate(labels):
        x, y = low_dim_embs[i, :]
        plt.scatter(x, y)
        plt.annotate(label,
                     xy=(x, y),
                     xytext=(5, 2),
                     textcoords='offset points',
                     ha='right',
                     va='bottom')

    plt.savefig(filename)
    
```

```python
from matplotlib import font_manager, rc #TSNE 그림에서 한글이 깨지는 문제를 해결해준다.
font_fname = 'c:/windows/fonts/H2GTRM.TTF'     # A font of your choice
font_name = font_manager.FontProperties(fname=font_fname).get_name()
rc('font', family=font_name)
```

```python
embedding_weights =new_d2v_model.wv.syn0
labels = new_d2v_model.wv.index2word
```

```python
tsne = TSNE(perplexity=30, n_components=2, init='pca', n_iter=5000)
plot_only = 50
low_dim_embs = tsne.fit_transform(embedding_weights[:plot_only, :])  
labels = [labels[i] for i in range(plot_only)]
plot_with_labels(low_dim_embs, labels)
```
