```python
from konlpy.tag import Kkma
```

```python
def tokenizing(content): #입력 문장 리스트를 토크나이징한다. Kkma는 pos로 분리하고, OS, NNG, NP만 빼내는 것이 좋다.
    kkma = Kkma()
    featurelist = []
    for term in kkma.pos(content):
        if term[1] == 'OS':
            featurelist.append(term[0])
        elif term[1] == 'NNG':
            featurelist.append(term[0])
        elif term[1] == 'NP':
            featurelist.append(term[0])
        elif term[1] == 'OL':
            featurelist.append(term[0])
        elif term[1] == 'NL':
            featurelist.append(term[0])
    return featurelist
```

```python
import gensim
import smart_open
```

```python
d2v_model = gensim.models.doc2vec.Doc2Vec(size=50, min_count=2, iter=55, dm=1, window=5) #파라미터를 다양하게 변화시켜 test해봐야합니다.
```

```python
d2v_wordlists = []
for doc in news:
    d2v_wordlists.append(gensim.models.doc2vec.TaggedDocument(words = doc["tokens"], tags = [doc["title"]['S']]))
```

```python
d2v_model.build_vocab(d2v_wordlists) #모델에 단어를 빌드len(d2v_termlists)for epoch in range(10):
```

```python
for epoch in range(10): #학습율을 조절하는 가장 일반적인 예시 코드
    d2v_model.train(d2v_wordlists,total_examples=d2v_model.corpus_count,epochs=d2v_model.iter)
    d2v_model.alpha -= 0.002 # decrease the learning rated2v_model.infer_vector(['in']) #입력 단어의 예측 벡터, 거의 0에서 머무른다.
    d2v_model.min_alpha = d2v_model.alpha # fix the learning rate, no deca
    d2v_model.train(d2v_wordlists,total_examples=d2v_model.corpus_count,epochs=d2v_model.iter)
```

```python
from scipy.spatial import distance
def searching(term, d2v): #단어와 각 문서간의 유클라디안 거리를 계산하는 함수
    d2v_dict = dict()
    for i in range(d2v.docvecs.count):
        d2v_dict[d2v.docvecs.index_to_doctag(i)] = distance.euclidean(d2v[term],d2v.docvecs[i])
        #print(d_key+" : "+str(distance.euclidean(d2v[term],d2v.docvecs[i])))
    return d2v_dict
```

```python
d2v_rank = searching('실적',new_d2v_model) #각 문서와 단어간에 유클라디안을 계산하여 저장
for i in sorted(d2v_rank, key=lambda k : d2v_rank[k]): #기술에 대한 각 문서별 유클라디안 거리를 오름차순으로 정렬한 모습.
    print( i + " : " + str(d2v_rank[i]))
```
